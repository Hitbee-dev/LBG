```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
units = 1000 # 개체 수
k = 10 # 최대 몇개까지 나눌건지
colors = ["black","red","coral","orange","gold","green","cyan","navy","gray","blueviolet"]
datax = [[],[],[],[],[],[],[],[],[],[]]
datay = [[],[],[],[],[],[],[],[],[],[]]
k_datax = [[],[],[],[],[],[],[],[],[],[]]
k_datay = [[],[],[],[],[],[],[],[],[],[]]

# 1000개의 개체를 랜덤으로 찍는다.
units_dot = np.random.rand(1000, 2)
plt.scatter(units_dot[:,0], units_dot[:,1], marker="o", s=20, c="gray")

# 1개의 centroid를 가운데 위치에 찍는다.
# for i in range()
k_dot = sum(units_dot[:,0])/len(units_dot[:,0]), sum(units_dot[:,1])/len(units_dot[:,1])
plt.scatter(k_dot[0], k_dot[1], marker="1", s=200, c="red")
plt.savefig(f'LBG0.png')
```

```python
# 랜덤으로 찍힌 1000개의 개체 중
plt.scatter(units_dot[:,0], units_dot[:,1], marker="o", s=20, c="gray")

for i in range(2):
    #가운데를 기준으로 세로로 반으로 나누고, 나눈 랜덤한 위치에 centroid를 찍는다.
    if(i == 0):
        k_datax[i] = random.uniform(min(units_dot[:,0]), k_dot[0])
    elif(i == 1):
        k_datax[i] = random.uniform(k_dot[0],max(units_dot[:,0]))
    k_datay[i] = random.uniform(min(units_dot[:,1]),max(units_dot[:, 1]))
    plt.scatter(k_datax[i], k_datay[i], marker="1", s=200, c="deeppink")
plt.savefig(f'LBG1.png')
```

```python
for i in range(units):
    best_distance = 10.0
    count = 0
    for j in range(2):
        distance = math.sqrt(pow(units_dot[i, 0]-k_datax[j], 2) + pow(units_dot[i, 1]-k_datay[j], 2))
        if(best_distance > distance):
            best_distance = distance
            count = j
    plt.scatter(units_dot[i, 0], units_dot[i, 1], marker="o", s=20, c=colors[count])
    datax[count].append(units_dot[i,0])
    datay[count].append(units_dot[i,1])

for i in range(2):
    plt.scatter(k_datax[i], k_datay[i], marker="1", s=200, c="deeppink")
plt.savefig(f'LBG2.png')
```

```python
for i in range(2):
    for j in range(len(datax[i])):
        plt.scatter(datax[i][j], datay[i][j], marker="o", s=20, c=colors[i])
for i in range(2):
    k_datax[i] = sum(datax[i])/len(datax[i])
    k_datay[i] = sum(datay[i])/len(datay[i])
    plt.scatter(k_datax[i], k_datay[i], marker="1", s=200, c="deeppink")
plt.savefig(f'LBG3.png')
```

```python
check = True
num = 3
while(check):
    num += 1
    cnt = 0
    rdatax = datax
    rdatay = datay
    datax = [[],[],[],[],[],[],[],[],[],[]]
    datay = [[],[],[],[],[],[],[],[],[],[]]

    for i in range(len(rdatax)): #10번 반복
        for j in range(len(rdatax[i])): #총 1000번 반복
            best_distance = 10.0
            count = 0
            for t in range(2):
                distance = math.sqrt(pow(rdatax[i][j]-k_datax[t], 2) + pow(rdatay[i][j]-k_datay[t], 2))
                if(best_distance > distance):
                    best_distance = distance
                    count = t
            plt.scatter(rdatax[i][j], rdatay[i][j], marker="o", s=20, color=colors[count])
            datax[count].append(rdatax[i][j])
            datay[count].append(rdatay[i][j])
    for i in range(2):
        if(k_datax[i] == sum(datax[i])/len(datax[i]) and k_datay[i] == sum(datay[i])/len(datay[i])):
            cnt += 1
        if(cnt == 2):
            check = False
    for i in range(2):
        k_datax[i] = sum(datax[i])/len(datax[i])
        k_datay[i] = sum(datay[i])/len(datay[i])
        plt.scatter(k_datax[i], k_datay[i], marker="1", s=200, color="deeppink")
    plt.savefig(f'LBG{num}.png')
    plt.show()
```
