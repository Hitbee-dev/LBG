```python
import matplotlib.pyplot as plt

import random
import lbg


N = 1000  # population size
SIZE_CODEBOOK = 10

random.seed(0)
p_x = []
p_y = []
cp_x = []
cp_y = []
# generate random Gauss distribution with µ=0, sigma=1
population = [(random.gauss(0, 1), random.gauss(0, 1))
              for _ in range(N)]

# display population as blue crosses
for i in population:
    p_x.append(i[0])
    p_y.append(i[1])
# plt.scatter(p_x,p_y,marker='o',color='gainsboro')
# plt.savefig('lbg2.png')
# plt.show()
##generate codebook
cb, cb_abs_w, cb_rel_w = lbg.generate_codebook(population, SIZE_CODEBOOK)
##display codebook as red filled circles
##codevectors with higher weight (more points near them) get bigger radius
colors = ["black","darkgray","rosybrown","maroon","orangered","saddlebrown","bisque","gold","yellow","greenyellow","lawngreen","forestgreen","cyan","deepskyblue","dodgerblue","royalblue","blueviolet","purple","fuchsia","hotpink","lightpink"]
for i in cb:
    cp_x.append(i[0])
    cp_y.append(i[1])
for i in range(len(cp_x)):
    for j in range(len(cb_abs_w[i])):
        plt.scatter(cb_abs_w[i][j][0],cb_abs_w[i][j][1],marker='o',color=colors[i])
    plt.scatter(cp_x[i],cp_y[i],s=200,marker='x',color=colors[0])
plt.savefig('lbg10.png')
plt.show()
```
